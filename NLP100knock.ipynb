## 09 Typoglycemia
スペースで区切られた単語列に対して，各単語の先頭と末尾の文字は残し，それ以外の文字の順序をランダムに並び替えるプログラムを作成せよ．ただし，長さが４以下の単語は並び替えないこととする．適当な英語の文（例えば”I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind .”）を与え，その実行結果を確認せよ．

```python
import random

def typoglycemia(str):
  words = str.split()
  result = ""
  for word in words:
    length = len(word)
    if length > 4:
      #mid = word[1:-1]
      shuffle = random.sample(word[1:length-1], length-2)   #random.sample(population, k)でランダムに並べた長さkのリストを作成 (random.shuffleも使える)
      word = word[0] + "".join(shuffle) + word[length-1]    #str.join(iterable)でリストなどの文字列を結合した文字列を形成
    result += word
    result += " "

  return result



print(typoglycemia("I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind ."))

#内包表記
str = "I couldn’t believe that I could actually understand what I was reading : the phenomenal power of the human mind ."
print(" ".join([word if len(word) <= 4 else word[0] +"".join(random.sample(word[1:len(word)-1], len(word)-2)) +  word[len(word)-1] for word in str.split()]))
```

## 20 JSONデータの読み込み
Wikipedia記事のJSONファイルを読み込み，「イギリス」に関する記事本文を表示せよ．問題21-29では，ここで抽出した記事本文に対して実行せよ．

```python
#!gunzip drive/MyDrive/'Colab Notebooks'/jawiki-country.json.gz

import json

def extract(country):
  fr = open("/content/drive/MyDrive/Colab Notebooks/jawiki-country.json", "r")
  for line in fr:
    data = json.loads(line)
    if data["title"] == country:
      return data["text"]
  fr.close()

print(extract("イギリス"))


fw = open("/content/drive/MyDrive/Colab Notebooks/jawiki-country_uk.json", "w")
fw.write(extract("イギリス"))
fw.close()

#with_open関数を使うと簡単にファイルの読み書きできる
```

## 21 カテゴリ名を含む行を抽出
記事中でカテゴリ名を宣言している行を抽出せよ．

```python
import json
import re

text = extract("イギリス")
categories = re.findall(r'\[\[Category:.*?\]\]', text)
for category in categories:
  print(category)
```

## 22 カテゴリ名の抽出
記事のカテゴリ名を（行単位ではなく名前で）抽出せよ．

```python
import json
import re

text = extract("イギリス")
categories = re.findall(r'\[\[Category:(.*?)(?:\|.*?)?\]\]', text) #正規表現は最終兵器として使う（処理速度遅い）
for category in categories:
  print(category)
```

## 23 セクション構造
記事中に含まれるセクション名とそのレベル（例えば”== セクション名 ==”なら1）を表示せよ．

```python
import json
import re

text = extract("イギリス")
sections = re.findall(r'(={2,})\s*(.+?)\s*(\1)', text)
for s in sections:
  print("{section}(レベル：{level})".format(section=s[1], level=len(s[0])-1))
```

## 26 強調マークアップの除去
25の処理時に，テンプレートの値からMediaWikiの強調マークアップ（弱い強調，強調，強い強調のすべて）を除去してテキストに変換せよ（参考: [マークアップ早見表](http://ja.wikipedia.org/wiki/Help:%E6%97%A9%E8%A6%8B%E8%A1%A8)）．

```python
import json
import re

color_dic = {'yellow':'\033[43m', 'red':'\033[31m', 'blue':'\033[34m', 'end':'\033[0m'}

def remove(str):
  result = re.sub(r"(\'{2,5})(.*?)(\1)", r"{}\2{}".format(color_dic['yellow'], color_dic['end']), str)
  return result

def print_dif(l):
  for fv in l:
      if fv[0] != remove(fv[0]) or fv[1] != remove(fv[1]):
        print("""--------除去前--------
{}: {}""".format(fv[0], fv[1]))
        print("""--------除去後--------
{}: {}\n""".format(remove(fv[0]), remove(fv[1])))

text = extract("イギリス")
temp = re.search(r"{{基礎情報.*?\n}}", text, re.DOTALL)
fields = re.findall(r"\|(.+?)\s*=\s*(.+?)(?=\n\|)", temp.group(), re.DOTALL)
#print({remove(f[0]):remove(f[1]) for f in fields})
print_dif(fields)
```

## 27 内部リンクの除去
26の処理に加えて，テンプレートの値からMediaWikiの内部リンクマークアップを除去し，テキストに変換せよ（参考: [マークアップ早見表](http://ja.wikipedia.org/wiki/Help:%E6%97%A9%E8%A6%8B%E8%A1%A8)）．

```python
import json
import re

def remove(str):
  tmp = re.sub(r"(\'{2,5})(.*?)(\1)", r"{}\2{}".format(color_dic['yellow'], color_dic['end']), str) #強調マークアップの除去
  result = re.sub(r"\[\[(?:[^|]*?\|)??([^|]*?)\]\]", r"{}\1{}".format(color_dic['red'], color_dic['end']), tmp) #内部リンクの除去
  return result

text = extract("イギリス")
temp = re.search(r"\{\{基礎情報.*?\n\}\}", text, re.DOTALL)
fields = re.findall(r"\|(.+?)\s*=\s*(.+?)(?=\n\|)", temp.group(), re.DOTALL)
#print({remove(f[0]):remove(f[1]) for f in fields})
print_dif(fields)
```

## 28 MediaWikiマークアップの除去
27の処理に加えて，テンプレートの値からMediaWikiマークアップを可能な限り除去し，国の基本情報を整形せよ．

```python
import json
import re

def remove(str):
  tmp = re.sub(r"(\'{2,5})(.*?)(\1)", r"\2", str) #強調マークアップの除去
  tmp = re.sub(r"\{\{lang\|([^|]*?\|)*?([^|]*?)\}\}", r"\2", tmp) #{{lang|言語名|説明文}}の除去
  tmp = re.sub(r"\[\[(?:[^|]*?\|)*?([^|]*?)\]\]", r"\1", tmp) #内部リンク, ファイルの除去
  tmp = re.sub(r"\[https?://[^>]*?\>([^\]]*?)\]", r" (\1)", tmp) #外部リンクの除去
  tmp = re.sub(r"\{\{仮リンク\|([^|]*?)(\|([^|]*?)){2}\}\}", r"\1", tmp)
  tmp = re.sub(r"\{\{Cite web\|url=([^|]*?)\|title=([^\|]*?)(\|([^|]*?))*?\}\}", r" (\2)", tmp)
  tmp = re.sub(r"<ref name=\"(.*?)\" />", r" (\1)", tmp)  #<ref name=>
  tmp = re.sub(r"\{\{0\}\}", "", tmp) #{{0}}
  tmp = re.sub(r"<.+?>", "", tmp) #<br />とか<ref></ref>
  return tmp

text = extract("イギリス")
temp = re.search(r"\{\{基礎情報.*?\n\}\}", text, re.DOTALL)
fields = re.findall(r"\|(.+?)\s*=\s*(.+?)(?=\n\|)", temp.group(), re.DOTALL)
dic = {remove(f[0]):remove(f[1]) for f in fields}
for f, v in dic.items():
  print("{}: {}".format(f,v))
#print(dic)
#print_dif(fields)
```

### 30 形態素解析結果の読み込み
形態素解析結果（neko.txt.mecab）を読み込むプログラムを実装せよ．ただし，各形態素は表層形（surface），基本形（base），品詞（pos），品詞細分類1（pos1）をキーとするマッピング型に格納し，1文を形態素（マッピング型）のリストとして表現せよ．第4章の残りの問題では，ここで作ったプログラムを活用せよ．

```python
#import MeCab

def mecab2list(mecabed_line):
  mecab_list = []
  lines = mecabed_line.splitlines()
  for l in lines:
    keitaiso = l.split("\t")
    if keitaiso[0] != "EOS":
      surface = keitaiso[0]
      base = keitaiso[3]
      pos_list=keitaiso[4].split("-")
      pos = pos_list[0]
      if len(pos_list)!=1:
        pos1 = pos_list[1]
      else:
        pos1 = ""
      if "記号" not in pos:
        map = {'surface': surface, 'base': base, 'pos': pos, 'pos1': pos1}
        mecab_list.append(map)
  return mecab_list

def read_mecab_text():
  result = []
  mecab = ""
  with open("/content/drive/MyDrive/Colab Notebooks/neko.txt.mecab", "r") as f:
    for line in f:
      if line.strip("\n")!="EOS":
        mecab += line
      else:
        mecab += line.strip("\n")
        result.append(mecab2list(mecab))
        mecab = ""
  return result

print(read_mecab_text()[1])
```

```python
import texthero as hero

"""
ストップワードとは、自然言語処理において分析対象から除外される単語のことです。
日本語では、「は」「の」「です」など、英語では「a」「the」「of」などは頻繁に用いられますが、
あまりに一般的な単語で分析対象としない場合が多く、その場合前処理などで除外することになります。
"""

def preprocess(raw_text):
  clean_text = hero.clean(raw_text, pipeline=[
      hero.preprocessing.lowercase, # 小文字への統一
      hero.remove_whitespace, # スペースの削除
      hero.preprocessing.remove_digits, # 数字の削除
      hero.preprocessing.remove_punctuation, # 句読点の削除
      hero.preprocessing.remove_diacritics, # ダイアクリティカルマーク（発音区別符号。àやéなど）の削除
      hero.preprocessing.remove_stopwords, # ストップワードの除去
      hero.preprocessing.stem
  ])
  return clean_text

raw_text = train_txt['TITLE']
clean_text = preprocess(raw_text)
hero.visualization.wordcloud(clean_text, colormap='viridis', width=500, height=400, background_color='White')
```

```python
from sklearn.cluster import KMeans
import numpy as np

n_clusters=5

country_code = pd.read_csv('country-code.csv')
countries = country_code.iloc[:,1]

country_name = []
country_vec = []
for _country in countries:
  country = _country.replace(' ','_').split(',')[0]
  if country in model:
    country_name.append(country)
    country_vec.append(model[country])

kmeans_model = KMeans(n_clusters=n_clusters, random_state=0).fit(country_vec)

for cluster in range(n_clusters):
  print('cluster: {}\n{}\n'.format(cluster, np.array(country_name)[(kmeans_model.labels_==cluster)]))
```

```python
from pymagnitude import Magnitude
import numpy as np
import pandas as pd

magnitude_file = 'competition/GoogleNews-vectors-negative300.magnitude'
w2vmodel = Magnitude(magnitude_file)

#SWEM

def get_sentence_vector(x: str, mode='mean'):
  embeddings = [
    w2vmodel.query(word)
    for word in x.split()
  ]
  if mode=='mean':
    return np.mean(embeddings, axis=0)
  elif mode=='max':
    return np.max(embeddings, axis=0)
  else:
    print("正しいmodeを入力してください")
    return 0

def w2v_feature(X: pd.DataFrame, mode='mean'):
  features = np.stack(
    X.map(lambda x: get_sentence_vector(x, mode)).values
  )
  return features
```

```python
import optuna

# 最適化対象を関数で指定
def objective_model(trial):
  # チューニング対象パラメータのセット
  C = trial.suggest_loguniform('C', 1e-4, 1e4)

  # モデルの学習
  model = LinearSVC(C=C)                          
  model.fit(train_X, train_y)

  # 予測値の取得
  pred = model.predict(valid_X)

  # 正解率の算出
  valid_accuracy = accuracy_score(valid_y, pred)
  return valid_accuracy 
```

```python
# 最適化
study = optuna.create_study(direction='maximize')
study.optimize(objective_model, n_trials=50)

# 結果の表示
print('Best trial:')
trial = study.best_trial
print('  Value: {:.3f}'.format(trial.value))
print('  Params: ')
for key, value in trial.params.items():
  print('    {}: {}'.format(key, value))
```
